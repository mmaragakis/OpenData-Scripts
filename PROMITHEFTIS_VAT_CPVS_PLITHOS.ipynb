```python
from numpy import equal
import pandas as pd
import csv
import matplotlib.pyplot as plt


# =====================================
# ΑΡΧΕΙΟ ΕΙΣΟΔΟΥ
# =====================================
input_file = "simbaseis 2018 2019 2023 2025 SERV.xlsx"

# =====================================
# ΔΙΑΒΑΣΜΑ EXCEL
# =====================================
df = pd.read_excel(input_file)

# Καθαρισμός ονομάτων στηλών
df.columns = df.columns.str.strip()

# Μετατροπή σε string
df["cpvs"] = df["cpvs"].astype(str)
df["vatNumber"] = df["vatNumber"].astype(str)


# =====================================
# ΠΛΗΘΟΣ ΑΝΑ CPVS
# =====================================
cpvs_counts = (
    df.groupby("cpvs")
      .size()
      .reset_index(name="plithos")
)


# =====================================
# DONUT CHART - TOP 9 CPVs
# =====================================

top9_cpvs = (
    cpvs_counts
    .sort_values("plithos", ascending=False)
    .head(9)
)

fig, ax = plt.subplots(figsize=(12, 8))

wedges, texts = ax.pie(
    top9_cpvs["plithos"],
    labels=top9_cpvs["cpvs"],      # ονόματα CPVs δίπλα στα κομμάτια
    startangle=90,
    labeldistance=1.08
)

# Δημιουργία donut
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig.gca().add_artist(centre_circle)

# Legend
ax.legend(
    wedges,
    top9_cpvs["cpvs"],
    title="Πλήθος Συμβάσεων Έργων για τα Κυριότερα CPVs",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

ax.set_title(
    "Top 9 CPVs με το Μεγαλύτερο Πλήθος Συμβάσεων",
    fontsize=14,
    pad=20
)

plt.tight_layout()

chart_output = "TOP9_CPVS_DONUT.png"

plt.savefig(
    chart_output,
    dpi=300,
    bbox_inches="tight"
)

plt.close()

print(f"Δημιουργήθηκε: {chart_output}")

# =====================================
# ΑΠΟΘΗΚΕΥΣΗ CSV ΓΙΑ CPVS
# =====================================
cpvs_output = "PLITHOS_per_cpv SERV SIMBASEIS.csv"

cpvs_counts.to_csv(
    cpvs_output,
    index=False,
    sep=';',
    encoding="utf-8-sig",
    quoting=csv.QUOTE_MINIMAL
)

# =====================================
# ΠΛΗΘΟΣ ΑΝΑ vatNumber
# =====================================
vat_counts = (
    df.groupby("vatNumber")
      .size()
      .reset_index(name="plithos2")
)

# =====================================
# ΑΠΟΘΗΚΕΥΣΗ CSV ΓΙΑ vatNumber
# =====================================
vat_output = "PLITHOS_per_vat SERV SIMBASEIS.csv"

vat_counts.to_csv(
    vat_output,
    index=False,
    sep=';',
    encoding="utf-8-sig",
    quoting=csv.QUOTE_MINIMAL
)


print(f"Δημιουργήθηκε: {cpvs_output}")
print(f"Δημιουργήθηκε: {vat_output}")
```
