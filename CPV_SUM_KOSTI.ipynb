```python
import pandas as pd
import csv

# === Ρυθμίσεις αρχείων ===simbaseis 2018 2019 2023 2025 ERGA
input_file = "simbaseis 2018 2019 2023 2025 SERV.xlsx"
output_file = "SUM_TZIROI_per_cpv SERV SIMBASEIS.csv"

# === Διαβάζουμε το Excel ===
df = pd.read_excel(input_file)

# === Καθαρισμός ονομάτων στηλών ===
df.columns = df.columns.str.strip()

# === Μετατροπή κόστους σε αριθμό ===
df["totalCostWithoutVAT"] = pd.to_numeric(
    df["totalCostWithoutVAT"],
    errors="coerce"
)

# === Μετατροπή CPV σε string ===
df["cpvs"] = df["cpvs"].astype(str)

# === Ομαδοποίηση ανά CPV ===
result = (
    df.groupby("cpvs", as_index=False)["totalCostWithoutVAT"]
    .sum()
)
result["totalCostWithoutVAT"] = result["totalCostWithoutVAT"].round(2)

#print(result)

# === Μετονομασία στηλών ===
result.columns = ["cpvs", "totalCostWithoutVAT"]

# === Αποθήκευση CSV με quotes ===
result.to_csv(
    output_file,
    index=False,
    sep='\t',
    encoding="utf-8-sig",
    quoting=csv.QUOTE_MINIMAL
)

print(f"Το αρχείο δημιουργήθηκε CPV: {output_file}")
```
